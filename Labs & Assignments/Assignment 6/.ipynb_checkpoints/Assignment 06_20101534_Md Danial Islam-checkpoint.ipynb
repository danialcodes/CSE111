```python
class Student:
  ID = 0
  def __init__(self,name,department,age,cgpa):
    
    Student.ID +=1
    self.ID = Student.ID
    self.name = name
    self.department = department
    self.age = age
    self.cgpa = cgpa

  def get_details(self):
    print("ID: {}\nName: {}\nDepartment: {}\nAge: {}\nCGPA: {}".format(self.ID,self.name,self.department,self.age,self.cgpa))

  @classmethod
  def from_String(cls,string):
    info = string.split('-')
    new = Student(info[0],info[1],int(info[2]),float(info[3]))
    return new

#Write your code here for subtasks 1-6.
s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()

# Write the answer of subtask 5 here
sub_tast_5 = '''\nA class variable is a variable which is related to the class and not an instance,
which means class variable can be use or called without creating an instance of the class, 
it is also remain same for all the class"s instances.
On the otherhand an instance variable is dependent on a particular instance, 
They can"t be called without creating an instance of the class.'''
print(sub_tast_5)
# Write the answer of subtask 6 here
sub_tast_6 = '''\nInstance method is dependent on the instances of the class,they take 'self' as compulsory parameter,
they can be called only by the referance of an instance of the class. They don't need any decorator to be made
On the other hand, class method need decorator @classmethod to be made, they take 'cls' as compulsory parameter,
also they can be called without creating an instance.'''
print(sub_tast_6)


#You are not allowed to change the code above
```

```python
class Assassin:
  total_assasin = 0

  def __init__(self,name,success_rate):
    Assassin.total_assasin +=1
    self.name = name
    self.success_rate = success_rate
  @classmethod
  def failureRate(cls,name,failureRate):
    return cls(name,100-failureRate)
  @classmethod
  def failurePercentage(cls,name,failurePercentage):
    return cls(name,100-failurePercentage)

  def printDetails(self):
    print('Name: {}\nSuccess rate: {}%\nTotal number of Assassin: {}'.format(self.name,self.success_rate,Assassin.total_assasin))

# Write your code here

john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
class Passenger:
  count = 0
  @classmethod
  def calculate_fare(cls,weight):
    if (weight<=20):
      return 0
    elif (weight<=50):
      return 50
    else:
      return 100
  def __init__(self,name):
    Passenger.count += 1
    self.name = name
    self.fare = 450
  def set_bag_weight(self,weight):
    self.weight = weight
    self.fare += Passenger.calculate_fare(weight)
  def printDetail(self):
    print('Name: {}\nBus Fare: {} taka'.format(self.name,self.fare))
  
# Write your code here

print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
class Travel:
  count = 0
  def __init__(self,source,destination):
    Travel.count +=1
    self.source = source
    self.destination = destination
    self.time = '1:00'
  def display_travel_info(self):
    return 'Source: {}\nDestination:{}\nFlight Time:{}'.format(self.source,self.destination,self.time)
  def set_time(self,time):
    self.time = str(time)+':00'
  def set_destination(self,destination):
    self.destination = destination
  def set_source(self,source):
    self.source = source
  

# Write your code here
print("No. of Traveller =", Travel.count)
print("=======================")
t1 = Travel("Dhaka","India")
print(t1.display_travel_info())
print("=======================")
t2 = Travel("Kuala Lampur","Dhaka")
t2.set_time(23)
print(t2.display_travel_info())
print("=======================")
t3 = Travel("Dhaka","New_Zealand")
t3.set_time(15)
t3.set_destination("Germany")
print(t3.display_travel_info())
print("=======================")
t4 = Travel("Dhaka","India")
t4.set_time(9)
t4.set_source("Malaysia")
t4.set_destination("Canada")
print(t4.display_travel_info())
print("=======================")
print("No. of Traveller =", Travel.count)
```

```python
from datetime import date

class Employee:
  @classmethod
  def employeeByJoiningYear(cls,name,year):
    today = date.today()
    return cls(name,(today.year-year))
  
  @staticmethod
  def experienceCheck(workingPeriod,gender):
    msg = 'is experienced'
    
    if workingPeriod < 3:
      msg = 'is not experienced'
    if gender=='female':
      return 'She '+msg
    
    return 'He '+msg
  
  def __init__(self,name,workingPeriod):
    self.name = name
    self.workingPeriod = workingPeriod
  


# Write your code here
employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
class Laptop:
  laptopCount=0
  
  @classmethod
  def addCount(cls,count):
    cls.laptopCount += count

  @classmethod
  def resetCount(cls):
    cls.laptopCount = 0

  @staticmethod
  def advantage():
    print('Laptops are portable')
  
  def __init__(self,name,count):
    self.name = name
    self.count = count
    Laptop.addCount(count)
  


# Write your code here
lenovo = Laptop("Lenovo", 5);
dell = Laptop("Dell", 7);
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
class Cat:
  Number_of_cats = 0
  @classmethod
  def no_parameter(cls,color='White',doing='sitting'):
    return cls(color,doing)
  @classmethod
  def first_parameter(cls,color):
    return cls.no_parameter(color=color)
  @classmethod
  def second_parameter(cls,doing):
    return cls.no_parameter('Grey',doing)
  @classmethod
  def addCat(cls):
    cls.Number_of_cats += 1

  def __init__(self,first_parameter,second_parameter):
    Cat.addCat()
    self.color = first_parameter
    self.doing = second_parameter
  def printCat(self):
    print('{} cat is {}'.format(self.color,self.doing))
  def changeColor(self,color):
    self.color = color
# Write your code here

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
import math

class Cylinder:
  radius = 5
  height = 18

  @staticmethod
  def area(r,h):
    total_area = 2 * math.pi * r *(r+h)
    print('Area: {}'.format(total_area))
  @staticmethod
  def volume(r,h):
    total_volume = math.pi * r * r * h
    print('Volume: {}'.format(total_volume))

  @classmethod
  def swap(cls,h,r):
    return cls(r,h)
  @classmethod
  def changeFormat(cls,input):
    input = input.split('-')
    instance = cls(float(input[0]),float(input[1]))
    return instance
  @classmethod
  def update_class_variable(cls,r,h):
    cls.radius = r
    cls.height = h
  
  def __init__(self, r = radius, h = height):
    self.radius = r
    self.height = h
    print('Default radius={} and height={}.\nUpdated: radius={} and height={}.'.
          format(Cylinder.radius,Cylinder.height,self.radius,self.height))
    Cylinder.update_class_variable(r,h)
# Write your code here

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
class Student:
  total_students = 0
  bracu_students = 0
  other_students = 0
  @classmethod
  def createStudent(cls,name,dept,institution='BRAC University'):
    return cls(name,dept,institution)
  @classmethod
  def printDetails(cls):
    print('Total Student(s): {}\nBRAC University Student(s): {}\nOther Institution Student(s): {}'.
          format(cls.total_students,cls.bracu_students,cls.other_students))
  @classmethod
  def updateCount(cls,institution):
    cls.total_students+=1
    if (institution=='BRAC University'):
      cls.bracu_students+=1
    else:
      cls.other_students+=1
    
  def __init__(self,name,dept,institution='BRAC University'):
    self.name = name
    self.dept = dept
    self.institution = institution
    Student.updateCount(institution)
    
  def individualDetail(self):
    print('Name: {}\nDepartment: {}\nInstitution: {}'.format(self.name,self.dept,self.institution))


# Write your code here

Student.printDetails()
print('#########################')
mikasa = Student('Mikasa Ackerman', "CSE")
mikasa.individualDetail()
print('------------------------------------------')
Student.printDetails()
print('========================')
harry = Student.createStudent('Harry Potter', "Defence Against Dark Arts", "Hogwarts School")
harry.individualDetail()
print('-------------------------------------------')
Student.printDetails()
print('=========================')
levi = Student.createStudent("Levi Ackerman", "CSE")
levi.individualDetail()
print('--------------------------------------------')
Student.printDetails()
```

```python
class SultansDine:
  branchCount = 0
  sellCount = 0
  branches_info =[]
  @classmethod
  def details(cls):
    print('Total Number of branch(s): {}\nTotal Sell: {} Taka'.format(cls.branchCount,cls.sellCount))
    if (cls.branchCount>0):
      for item in cls.branches_info:
        print("Branch Name: {}, Branch Sell: {} Taka\nBranch consists of total sell's: {:.2f}% ".format(item[0],item[1],item[2]))

  @classmethod
  def update(cls,name,sell):
    cls.branchCount+=1
    cls.sellCount += sell
    sell_percentage = (sell/cls.sellCount)*100
    check = 0
    if ([name,sell,sell_percentage] not in cls.branches_info):
      cls.branches_info.append([name,sell,sell_percentage])

    for item in cls.branches_info:
      index = cls.branches_info.index(item)
      sell_percentage = (item[1]/cls.sellCount)*100
      cls.branches_info[index][2] = sell_percentage
      
  def __init__(self,name):
    self.b_name = name
    self.sell = 0

  def sellQuantity(self,quantity):
    if (quantity < 10):
      unit_price = 300
    elif (quantity < 20):
      unit_price = 350
    else:
      unit_price = 400
    self.sell += quantity * unit_price
    SultansDine.update( self.b_name , self.sell)

  def branchInformation(self):
    print('Branch Name: {}\nBranch Sell: {} Taka'.format(self.b_name,self.sell))



# Write your code here
SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
