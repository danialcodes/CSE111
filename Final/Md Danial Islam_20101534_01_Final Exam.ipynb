```python
#Task_1
user_input = input()
num = int(input())
dic = {}
for i in range(len(user_input)):
  word = user_input[i:i+num]
  if (word in dic):
    dic[word]+=1
  else:
    dic[word] = 1
  if (i == (len(user_input) - num)):
    break
highest = [0,0]
count = 0
for k,v in dic.items():
  print(k,v)
  if (v > highest[1]):
    highest = [k,v]
  elif (v == highest[1]):
    count+=1
if (count>=2):
  print("No highest substring found.")
else:
  print("Highest substring: {}".format(highest[0]))

```
